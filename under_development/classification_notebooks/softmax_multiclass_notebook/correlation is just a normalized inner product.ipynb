I've found that some people don't have an intuitive grasp as to why correlation is a measurement of linearity between random variables, and why this value is between 1 and -1, so I want to try to fill in the gap. These properties are traceable to the fact that correlation may be written as a normalized inner product (also called a dot product in the sample case).

Lets illustrate the idea using perhaps the greatest discovered sample correlation in the history of the universe, captured in this lovely poster designed by the brilliant Jeffery Frankenhauser at frankenspace.com (click to enlarge).

Anyway, lets denote by $\mathcal{X}=\left\{ x_{1}...x_{10}\right\}$ and $\mathcal{Y}=\left\{ y_{1}...y_{10}\right\}$ the sample data for the number of explosions and gross profit respectively for the Michael Bay "movies" listed above. To simplify notation assume I've subtracted the mean from both sample sets. I'm ball-parking all the numbers using the image so things won't be precise, but you'll get the general point. The sample correlation between $\mathcal{X}$ and $\mathcal{Y}$ (turns out to be 0.96) is calculated

\begin{equation}
r_{\mathcal{X},\mathcal{Y}}=\frac{Cov\left(\mathcal{X},\mathcal{Y}\right)}{\sqrt{Var\left(\mathcal{X}\right)Var\left(\mathcal{Y}\right)}}=\frac{\frac{1}{9}\underset{i=1}{\overset{10}{\sum}}\left(x_{i}y_{i}\right)}{\sqrt{\left(\frac{1}{9}\underset{i=1}{\overset{10}{\sum}}x_{i}^{2}\right)\left(\frac{1}{9}\underset{i=1}{\overset{10}{\sum}}y_{i}^{2}\right)}}=\frac{\mathbf{x}\cdot\mathbf{y}}{\Vert\mathbf{x}\Vert\Vert\mathbf{y}\Vert}
\end{equation}

Here I've just re-written the sum/products using vector notation where $\mathbf{x}=\left[x_{1}...x_{10}\right]$ and  $\mathbf{y}=\left[y_{1}...y_{10}\right]$.

Does the right hand side look familar?

The inner product - a blast from the past

Take two vectors $\mathbf{x}$ and $\mathbf{y}$ in $\mathbb{R}^{N}$, then the inner product (a generalization of multiplication to higher dimensions) between them is defined $\mathbf{x}\cdot\mathbf{y}=\underset{i=1}{\overset{N}{\sum}}x_{i}y_{i}$. Remember that, through the magic of trigonometry, the angle $\theta$ between these vectors is related to their inner product by $\mathbf{x}\cdot\mathbf{y}=\Vert\mathbf{x}\Vert\Vert\mathbf{y}\Vert cos\left(\theta\right)$, and rearranged this gives the normalized inner product $\frac{\mathbf{x}\cdot\mathbf{y}}{\Vert\mathbf{x}\Vert\Vert\mathbf{y}\Vert}=cos\left(\theta\right)$
which measures how well $\mathbf{x}$ and $\mathbf{y}$ allign linearly, that is how well they point in the same direction. Here's a picture illustrating the concept where $\theta$ is solved for, you've probably seen this a thousand times in some previous life.

Now $cos\left(\theta\right)\in\left[-1\,1\right]$ and when

$cos\left(\theta\right)>0$ the two vectors are pointing into the same half-space of $\mathbb{R}^{N}$ and when $ cos\left(\theta\right)=1$ they perfectly alligned (since $\theta=0$).  Conversely when $ cos\left(\theta\right)<0$ the two vectors point into disjoint halfspaces and when $ cos\left(\theta\right)=-1$ they point in exactly opposite directions (since $ \theta$ is an odd multiple of $ \pi$).  When $ \mathbf{x}$ and $\mathbf{y}$ are thought of as a vectors of single dimensional observations $ cos\left(\theta\right)=\pm1$ implies $ \mathbf{x}=\pm C\mathbf{y}$ for a constant $ C>0$, e.g. that the observations fall on a straight line in the plane. The closer is $ cos\left(\theta\right)$ to $ \pm1$ the more strictly do the observations lie on a line.
$ cos\left(\theta\right)=0$ the two vectors are orthogonal since $ \theta$ is an odd multiple of $ \frac{\pi}{2}$.  This implies the observations (entries of the vectors) are uncorrelated.
The inner product viewpoint on correlation provides the intuitive reason why correlation is said to measure the strength of the linear relationship between two sets of observations, and that this measurment lies in the range $ \left[-1\,1\right]$.  This is because the inner product is a measurement of how well aligned the observation vectors are, or how well the point in the same ( linear) direction: if they point in exactly the same direction the correlation is 1, if they point in perpendicular directions the correlation is 0, if the point in exactly opposite directions the correlation is -1, etc.

Correlation involving distributions

Now lets examine the general definition of correlation when distributions are involved. To see that a normalized inner product is at play, write out the correlation of two zero-mean continuous random variables $latex X$ and $latex Y$ with densities $latex f$ and $latex g$ respectively. That is

$latex \rho_{X,Y}=\frac{Cov\left(X,Y\right)}{\sqrt{Var\left(X\right)Var\left(Y\right)}}=\frac{E\left[XY\right]}{\sqrt{Var\left(X\right)Var\left(Y\right)}}=\frac{\int\int xyf\left(x\right)g\left(y\right)dxdy}{\sqrt{\left(\int x^{2}f\left(x\right)dx\right)\left(\int y^{2}g\left(y\right)dy\right)}}$

Compare this to the sample correlation of two sets of $latex M$ samples $latex \mathcal{X}$ and $latex \mathcal{Y}$ restated here slightly more generally

$latex r_{\mathcal{X},\mathcal{Y}}=\frac{\frac{1}{M-1}\underset{i=1}{\overset{M}{\sum}}\left(x_{i}y_{i}\right)}{\sqrt{\left(\frac{1}{M-1}\underset{i=1}{\overset{M}{\sum}}x_{i}^{2}\right)\left(\frac{1}{M-1}\underset{i=1}{\overset{M}{\sum}}y_{i}^{2}\right)}}$

They look pretty similar right? Replace the sums with integrals, and the constant probability $latex \frac{1}{M-1}$ with $latex f$ and $latex g$ respectively and you're there - and indeed as $latex X\cdot Y=E\left[XY\right]$ is an inner product in the same ways $latex \mathbf{x}\cdot\mathbf{y}$ is.  You can think about this as a dot product of infinite dimensional vectors.
