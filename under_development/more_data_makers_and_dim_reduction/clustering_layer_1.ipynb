```python
### in this cell we import necessary libraries for the demo
import numpy as np                  # a fundamental numerical linear algebra library
import matplotlib.pyplot as plt     # a basic plotting library
import pandas as pd                 # a basic data handling library
import sys
sys.path.append('utils')

# this line is required in order to plot in a jupyter notebook itself
%matplotlib inline         

# load in the data 
import clustering_layer1_utils as utils  # a set of simple plotting utilities for this notebook
data_x,data_y,labels = utils.load_data('test_data.csv')

# plot the original clusters
utils.plot_orig_data(data_x,data_y,labels)
```

```python
# Kmeans is a clustering algorithm from the scikit-learn library
from sklearn.cluster import KMeans

# create instance of clustering algo
clf = KMeans(n_clusters=3)

# cluster dataa using our chosen algo
data = np.concatenate((data_x,data_y),axis = 1)
clf.fit(data)

# plot
utils.plot_clustered_data(data_x,data_y,clf)
```
