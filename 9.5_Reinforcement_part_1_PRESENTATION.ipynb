-  Three main application areas in which RL is commonly used: robotics, automatic control, and game AI.

# Principles of Reinforcement  

# Principles of Reinforcement  

- **Principle 1:** Reinforcement Learning automates narrowly focused goal-driven tasks.    Narrowly focused $\neq$ simple.  <p> <img src="demo_images/RL_images/maze.gif" alt="Smiley face" width="200" height="200" align="right"> </p>

- **Principle 2:**  Experimentation in the space of a problem - commonly a realistic simulation of a problem's environment -  is crucial to training a Reinforcement Learning agent.

- **Principle 3:** A Reinforcement Learning agent learns by repeated trial-and-error interactions with a problem environment or simulator. <p> <img src="demo_images/RL_images/gridworld_training.gif" width="800" align="middle"> </p>
    


- **Principle 4:** Numerical rewards are used to communicate the desired goal to the Reinforcement Learning agent.

# The fundamental components of an RL problem

# The fundamental components of an RL problem

1\.    A **state** is a specific configuration of all objects in a problem's environment
    - Defined by environment

2\.  The agent can then take an **action** 
    - Defined by environment

3\.  A **reward** is a numerical signal communicating our desired goal to the agent
    - Defined by we humans

## Example: Gridworld

- A **state**  current location of the player (the blue square). 

- **Actions** the agent can one square adjacently up, down, left or right.  

- **Rewards** 
    - -1 to all actions leading to non-goal/non-hazard
    - -100 for those leading to hazard
    - 0 for actions leading to goal


## Example: Chess

- A **state** is any (legal) configuration of the white and black pieces.

- An **action** is any legal move of current piece on the board.


- **Rewards** 
    - action leading to checkmate gives 10000
    - every other action gives -1

## Example:  Inverted pendulum

 - a **state** is a complete set of information about the cart and pole's position

- An **action** is moving the cart left or right along the horizontal axis by one unit.


- **Rewards** 
    - angle between the pole and the horizontal axis is above a certain threshold the reward 1
    - otherwise reward is 0

# Formalities and jargon of Reinforcement

# Modeling and jargon of Reinforcement

- **Principle 3:** A Reinforcement Learning agent learns by repeated trial-and-error interactions with a problem environment or simulator. <p> <img src="demo_images/RL_images/gridworld_training.gif" width="600" height="600" align="middle"> </p> 

1\.  Run through many **simulations**

2\.  Each simulation run is called an **episode**

3\. Each movement is called a **step**

<p> <img src="demo_images/RL_images/gridworld_training.gif" width="900" align="middle"> </p>

4\. Each step consists of 4 events: 
    - begin at a state
    - take an action
    - move to a new state based on action taken
    - receive reward based on action taken

# Formal modeling: an episode of steps

# Formal modeling: an episode of steps

**step 2:**<br>
- start at state $s_1$<br>
- take an action (denoted by $a_2$)<br>
- arrive at new state (denoted by $s_2$) and receive second reward $r_2$<br>


**step 3:**<br>
- start at state $s_2$<br>
- take an action (denoted by $a_3$)<br>
- arrive at new state (denoted by $s_3$) and and receive third reward $r_3$<br>



**step 4:**<br>
...<br>

Written more compactly as:

($s_0$, $a_1$, $r_1$), ($s_1$, $a_2$, $r_2$), ($s_2$, $a_3$, $r_3$), ($s_3$,...


# Step back: what are we after again?

- simulations $\longrightarrow$ episodes $\longrightarrow$ steps 

($s_0$, $a_1$, $r_1$), ($s_1$, $a_2$, $r_2$), ($s_2$, $a_3$, $r_3$), ($s_3$,...


- want: **optimal actions leading to greatest long-term reward**

# Reinforcement fundamentals: the Q cost function

# The Q cost function

- want: **optimal actions leading to greatest long-term reward** 

- expressing this desire in math gives us so-called Q cost 

- derivation uses common math trick

- assume what you want, use it to derive (**recursive equations**)

## The Q cost function: derivation

- take a sequence of steps:

($s_0$, $a_1$, $r_1$), ($s_1$, $a_2$, $r_2$), ($s_2$, $a_3$, $r_3$), ($s_3$,...

- start at $s_0$ 
$\longrightarrow$ take any first action $a_1$ $\longrightarrow$  recieve reward $r_1$ 

-  **now assume every further action is optimal**

- denote the resulting maximal long-term reward: $Q(s_0,a_1)$

## The Q cost function: derivation

- denote the resulting maximal long-term reward by $Q(s_0,a_1)$

- only action **not** potentially optimal is first at $s_0$ i.e., $a_1$

- then maximum total reward from $s_0$ :   $\underset{i}{\text{maximum}}\,\,Q\left(s_{0},\,\alpha_{i}\right)$


## The Q cost function: derivation

- then maximum total reward from $s_0$ is:   $\underset{i}{\text{maximum}}\,\,Q\left(s_{0},\,\alpha_{i}\right)$

- with this $Q$ is defined *recursively* in terms of itself

- $Q(s_0,a_1)$ must be the sum of $r_1$ **and** the *maximum possible sum of rewards for all subsequent steps:* $\underset{i}{\text{maximum}}\,\,Q\left(s_{1},\,\alpha_{i}\right)$.

- Algebraically:   $Q\left(s_{0},\,a_{1}\right)=r_{1}+\underset{i=1}{\text{maximum}}\,\,Q\left(s_{1},\,\alpha_{i}\right)$

## The Q cost function: derivation

So can approximate $Q$ by running through an episode of simulation 

($s_0$, $a_1$, $r_1$), ($s_1$, $a_2$, $r_2$), ($s_2$, $a_3$, $r_3$), ($s_3$,...

and updating $Q$ recursively at each step:

$Q\left(s_{0},\,a_{1}\right)=r_{1}+\underset{i=1...M}{\text{maximum}}\,\,Q\left(s_{1},\,\alpha_{i}\right)$

$Q\left(s_{1},\,a_{2}\right)=r_{2}+\underset{i=1...M}{\text{maximum}}\,\,Q\left(s_{2},\,\alpha_{i}\right)$

$Q\left(s_{2},\,a_{3}\right)=r_{3}+\underset{i=1...M}{\text{maximum}}\,\,Q\left(s_{3},\,\alpha_{i}\right)$

$\vdots$

$Q\left(s_{k-1},\,a_{k}\right)=r_{k}+\underset{i=1...M}{\text{maximum}}\,\,Q\left(s_{k},\,\alpha_{i}\right)$

$\vdots$

## The Q cost function: derivation

- This is called **the Q-Learning Algorithm**

- works by approximating $Q$ iteratively

- updating $Q$ recursively at each step of each simulation

- exactly as shown here
<p>
<img src="demo_images/RL_images/gridworld_training.gif" width="800" align="middle"> </p>

### Q -learning pseudo-code
Initialize $Q$

**for** e = 1...E (the maximum number of episodes)
    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Select a random initial state $s_0$ and set $k=1$
    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; **while** end state not reached AND maximum iteration count not met

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;select the action $a_k$ at random and record the resulting state $s_k$ and corresponding reward $r_k$

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;update $Q$ as 
    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$Q\left(s_{k-1},\,a_{k}\right)=r_{k}+\underset{i=1...M}{\text{maximum}}\,\,Q\left(s_{k},\,\alpha_{i}\right)$
    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$k \longleftarrow k+1$

# Q-learning for the gridworld maze

```python
### start main Q-learning loop ###
for n in range(self.training_episodes): 
    # pick this episode's starting position
    grid.agent = self.training_start_schedule[n]

    # update Q matrix while loc != goal
    episode_history = []      # container for storing this episode's journey
    total_episode_reward = 0
    for step in range(self.max_steps):   
        # update episode history container
        episode_history.append(grid.agent)

        ### if you reach the goal end current episode immediately
        if grid.agent == grid.goal:
            break

        # translate current agent location tuple into index
        s_k_1 = grid.state_tuple_to_index(grid.agent)

        # get action
        a_k = grid.get_action(method = self.action_method,Q = Q,exploit_param = self.exploit_param)

        # move based on this action
        s_k = grid.get_movin(action = a_k)

        # get reward     
        r_k = grid.get_reward(state_index = s_k)          

        # update Q
        Q[s_k_1,a_k] = r_k + max(Q[s_k,:])

        # update current location of agent 
        grid.agent = grid.state_index_to_tuple(state_index = s_k)
```

## Applying Q-learning to the gridworld maze

```python
# import q-learning module
from gridworld_lib import gridworld_qlearn

# create an instance of the q-learner
qlearner = gridworld_qlearn.learner(gridworld = small_maze,action_method = 'random')

# run q-learning
qlearner.train(verbose = True) 
```
