```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s = np.linspace(0,1,200)
t = np.exp(3*s)*(np.sin(3*np.pi**2*(s - 0.5)))/(3*np.pi*(s-0.5))
plt.plot(s,t)
s.shape = (len(s),1)
t.shape = (len(t),1)
data = np.concatenate((s,t),axis = 1)
```

```python
np.savetxt('guess_this.txt',data,delimiter=',')
```
